```python
import yfinance as yf
import statsmodels.api as sm
from datetime import timedelta
from datetime import datetime
```

```python
acao = input("For which BR stock would you like to calculate the regression beta? ").upper()
acao = acao + ".SA"

tickers = [acao, "^BVSP"]

data = datetime.now()

tres_anos_atras = data - timedelta(days = 1095)

dados_cotacoes = yf.download(tickers, tres_anos_atras, data)['Adj Close']

dados_cotacoes
```

```python
retornos_diarios = dados_cotacoes.pct_change().dropna()

retornos_diarios
```

```python
X = retornos_diarios['^BVSP']
Y = retornos_diarios[tickers[0]]
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
```

```python
# Beta 

print(model.params[1])
print(model.rsquared)
print("--------------------------------")
print(str(model.summary()))
```
