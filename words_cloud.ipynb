```python
from pypdf import PdfReader
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
reader = PdfReader("klabin.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
conectivos = set(STOPWORDS)
conectivos.update(["da", "meu", "em", "e", "você", "de", "ao", "os", 'na', 'o', '2T24', 'para', 'milhões',
                  'resultado', 'que', 'nas', 'dos', "n", "p", "1T24", "foi", 'ano', 'Klabin', '4T23', 'resultados',
                  'trimestre', 'weg', '2T23'])

```

```python
nuvem_palavras = WordCloud(stopwords= conectivos, background_color= 'black', 
                           width = 1920, height = 1080).generate(text)
```

```python
fig, ax = plt.subplots()
ax.imshow(nuvem_palavras)
ax.set_axis_off()
```
